## Q1. What are the three measures of central tendency?

__The three measures of central tendency are:__

__Mean__: The arithmetic average of a dataset.

__Median__: The middle value of a dataset when it is ordered.

__Mode__: The most frequently occurring value in a dataset.

## Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

__Mean__: Calculated by summing all the values and dividing by the number of values. It is used when data are symmetrically distributed without outliers.

__Median__: The middle value when the data are arranged in ascending order. It is used when the data are skewed or have outliers.

If the number of observations (N) is odd, the median is the middle number. If N is even, the median is the average of the two middle numbers.

__Mode__: The value that appears most frequently. It is used for categorical data or to understand the most common occurrence in the data.

## Q8. What do you understand about skewness in data?

Skewness refers to the asymmetry in the distribution of data. It indicates whether the data values are skewed to the left (negative skew) or right (positive skew):

__Positive Skew:__ Tail on the right side, mean > median.

__Negative Skew:__ Tail on the left side, mean < median.

## Q9. If a data is right-skewed then what will be the position of median with respect to mean?

__In a right-skewed (positively skewed) data, the median will be less than the mean.__

## Q12. For a normal distribution data what is the relationship between its measure of central tendency?

__In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.__

## Q13. How is covariance different from correlation?

__Covariance indicates the direction of the linear relationship between variables but does not provide the strength of the relationship in a standardized way. Correlation, however, standardizes the covariance, providing a dimensionless value that indicates both the strength and direction of the relationship.__



## Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

__Central Tendency:__ Outliers can significantly affect the mean, pulling it towards the outlier. The median and mode are less affected by outliers.

__Example:__ For the dataset [1, 2, 3, 4, 100], the mean is 22, which is heavily influenced by the outlier 100, while the median is 3, reflecting the central tendency better.

__Dispersion:__ Outliers increase measures of dispersion such as range, variance, and standard deviation, making the data appear more spread out.

__Example:__ For the same dataset [1, 2, 3, 4, 100], the standard deviation is much higher than for [1, 2, 3, 4, 5].
